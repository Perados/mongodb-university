```python
import pandas as pd
import pymongo
```

```python
connection = pymongo.MongoClient('mongodb://localhost')
```

```python
db = connection.students
```

```python
student_ids = db.grades.distinct('student_id')
```

```python
def remove_lowest_score(db, student_id):
    object_id = list(
        db.grades.find({'student_id': student_id, 'type': 'homework'}).sort('score', pymongo.ASCENDING)
    )[0]['_id']
    print(object_id)
    result = db.grades.delete_many({'_id': object_id})
    print(result.deleted_count)
```

```python
for id in student_ids:
    remove_lowest_score(db, id)
```
