```python
import pymongo
```

```python
connection = pymongo.MongoClient('mongodb://localhost')
```

```python
db = connection.school
```

```python
cursor = db.students.find()
```

```python
students = list(cursor)
```

```python

for student in students:
    scores = student['scores']
    homeworks = [score for score in scores if score['type'] == 'homework']
    lowest_score = sorted(homeworks, key=lambda k: k['score'])[0]
    new_scores = [score for score in scores if score != lowest_score]
    print 'scores', scores
    print 'homeworks:', homeworks
    print 'lowest:', lowest_score
    print 'new:', new_scores
    db.students.update_one({'_id': student['_id']}, {'$set': {'scores': new_scores}})
    
```
